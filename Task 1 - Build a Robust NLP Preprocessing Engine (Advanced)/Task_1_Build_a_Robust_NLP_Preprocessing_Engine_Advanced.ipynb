```python
import re
from collections import Counter

all_processed_tokens = []
all_processed_sentences = []
```

```python
def preprocess_text(text):

    if not isinstance(text, str) or not text.strip():
        return [], ""

    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+|\S+@\S+', '', text)
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    exception_words = {'no', 'not'}
    clean_tokens = [word for word in tokens if len(word) > 2 or word in exception_words]

    clean_sentence = " ".join(clean_tokens)

    return clean_tokens, clean_sentence
```

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("--- Task 3: Stress Testing Outputs ---")
for text in sample_inputs:
    tokens, clean_sentence = preprocess_text(text)

    all_processed_tokens.append(tokens)
    all_processed_sentences.append(clean_sentence)

    print(f"Original Text   : {repr(text)}")
    print(f"Cleaned Tokens  : {tokens}")
    print(f"Cleaned Sentence: {repr(clean_sentence)}")
    print("-" * 40)
```

```python
print("--- Task 4: Token Analytics ---")
for i, tokens in enumerate(all_processed_tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"Sentence {i+1} Analytics:")
    print(f"  Total Tokens : {total_tokens}")
    print(f"  Unique Tokens: {unique_tokens}")
    print(f"  Avg Length   : {avg_length:.2f}")

print("\nTask 4 Analysis:")
print("- Most noise: Sentences like 'Win $$$ now!!! Limited offer!!!' had the most noise removed.")
print("- Most meaningful tokens: 'I absolutely looooved this product' retained the most meaning.")
```

```python
print("--- Task 5: Frequency Analysis ---")

flat_token_list = [token for sublist in all_processed_tokens for token in sublist]
token_counts = Counter(flat_token_list)

print("Top 10 Most Frequent Words:")
for word, count in token_counts.most_common(10):
    print(f"  {word}: {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in token_counts.most_common()[:-6:-1]:
    print(f"  {word}: {count}")
```

```python
def full_pipeline(text_list):
    pipeline_tokens = []
    pipeline_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        pipeline_tokens.append(tokens)
        pipeline_sentences.append(clean_sentence)

    return {
        "tokens": pipeline_tokens,
        "clean_sentences": pipeline_sentences
    }

pipeline_output = full_pipeline(["Test sentence one.", "Another test!"])
print(pipeline_output)
```

```python
print("--- Task 7: Error Handling Tests ---")

error_test_cases = [
    "",
    "😊😂👍",
    "12345 67890"
]

error_results = full_pipeline(error_test_cases)

for i, original in enumerate(error_test_cases):
    print(f"Original Text   : {repr(original)}")
    print(f"Cleaned Tokens  : {error_results['tokens'][i]}")
    print(f"Cleaned Sentence: {repr(error_results['clean_sentences'][i])}")
    print("-" * 40)
```
